#### Формування множини векторів сигма

```python
def compare_alternatives ( a1, a2):
    comparison_result = np.arange(len(a1))
    for i in range(0, len(a1)):
        if a1[i] > a2[i]:
            comparison_result[i] = 1
        elif a1[i] < a2[i]:
            comparison_result[i] = -1
        elif a1[i] == a2[i]:
            comparison_result[i] = 0
    return comparison_result

def sigma_vectors(matrix):
    alternatives = np.array(matrix)
    result_matrix = [[0]*20 for i in range(20)]
    for i in range(0, 20):
        for j in range(0, 20):
            result_matrix[i][j] = compare_alternatives(alternatives[i], alternatives[j])
            #print(compare_alternatives(alternatives[i], alternatives[j]))
    return result_matrix

sigma_vectors(task_array)
    
```

#### Максимальні та найбільші за P та R

```python
def I(r):
    return (r==r.T)*r

def dom_max_p(m):
    matrix = np.array(m)
    max_ = []
    for i in range(0, len(matrix)):
        if matrix[i][i]==0 and matrix[i].sum()==len(matrix)-1:
            max_.append(i)
    return max_

def dom_max_r(m):
    matrix = np.array(m)
    max_ = []
    strong_max = []
    for i in range(0, len(matrix)):
        if matrix[i].sum()==len(matrix):
            max_.append(i)
            if matrix[:,i].sum()==1:
                strong_max.append(i)
    return max_, strong_max

def block_max_p(m):
    matrix = np.array(m)
    max_ = []
    for i in range(0, len(matrix)):
        if matrix[:,i].sum()==0:
            max_.append(i)
    return max_

def block_max_r(m):
    matrix = np.array(m)
    sym = I(matrix)
    max_ = []
    strong_max = []
    for i in range(0, len(matrix)):
        if matrix[:,i]==sym[:,i]:
            max_.append(i)
            if matrix[:,i].sum()==1 and matrix[i][i]==1:
                strong_max.append(i)
    return max_, strong_max
    
```

#### Відношення Парето

```python
def pareto_check_vector(vector):
    check = np.array(vector)
    flag = True
    for i in range (0, len(check)):
        if check[i]>=0:
            pass
        else:
            flag = False
    if flag:
        return 1
    else:
        flag = True
        for i in range (0, len(check)):
            if check[i]<=0:
                pass
            else:
                flag = False
        if flag:
            return 2
        else:
            return 3
    
def pareto(sigma_vectors):
    vectors_matrix = np.array(sigma_vectors)
    result_matrix = [[0]*20 for i in range(20)]
    for i in range(0, 20):
        result_matrix[i][i] = 1
    for i in range(0, 20):
        for j in range(i+1, 20):
            flag = pareto_check_vector(vectors_matrix[i][j])
            if flag == 1:
                result_matrix[i][j] = 1
                result_matrix[j][i] = 0
            elif flag == 2:
                result_matrix[i][j] = 0
                result_matrix[j][i] = 1
            elif flag == 3:
                result_matrix[i][j] = 0
                result_matrix[j][i] = 0
    if len(dom_max_p(result_matrix))>0:
        print("Найбільший за Р: {}".format(dom_max_p(result_matrix)))
    else:
        print("Максимальний за Р: {}".format(block_max_p(result_matrix)))
    return result_matrix

pareto(sigma_vectors(task_array))
```
