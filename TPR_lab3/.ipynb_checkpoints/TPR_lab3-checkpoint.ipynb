#### Формування множини векторів сигма

```python
def compare_alternatives ( a1, a2):
    comparison_result = np.arange(len(a1))
    for i in range(0, len(a1)):
        if a1[i] > a2[i]:
            comparison_result[i] = 1
        elif a1[i] < a2[i]:
            comparison_result[i] = -1
        elif a1[i] == a2[i]:
            comparison_result[i] = 0
    return comparison_result

def sigma_vectors(matrix):
    alternatives = np.array(matrix)
    result_matrix = [[0]*20 for i in range(20)]
    for i in range(0, 20):
        for j in range(0, 20):
            result_matrix[i][j] = compare_alternatives(alternatives[i], alternatives[j])
    return result_matrix

s_v = sigma_vectors(task_array)
s_v
```

#### Максимальні та найбільші за P та R

```python
def I(r):
    return (r==r.T)*r


def dom_max_p(m):
    matrix = np.array(m)
    max_ = []
    for i in range(0, len(matrix)):
        if matrix[i][i]==0 and matrix[i].sum()==len(matrix)-1:
            max_.append(i)
    return max_

def dom_max_r(m):
    matrix = np.array(m)
    max_ = []
    strong_max = []
    for i in range(0, len(matrix)):
        if matrix[i].sum()==len(matrix):
            max_.append(i)
            if matrix[:,i].sum()==1:
                strong_max.append(i)
    return max_, strong_max

def block_max_p(m):
    matrix = np.array(m)
    max_ = []
    for i in range(0, len(matrix)):
        if matrix[:,i].sum()==0:
            max_.append(i)
    return max_

def block_max_r(m):
    matrix = np.array(m)
    sym = I(matrix)
    max_ = []
    strong_max = []
    for i in range(0, len(matrix)):
        if np.any(np.array_equal(matrix[:,i], sym[:,i])==False)==False:
            max_.append(i)
            if matrix[:,i].sum()==1 and matrix[i][i]==1:
                strong_max.append(i)
    return max_, strong_max
    
```

#### Відношення Парето

```python
def pareto_check_vector(vector):
    check = np.array(vector)
    flag = True
    for i in range (0, len(check)):
        if check[i]>=0:
            pass
        else:
            flag = False
    if flag:
        return 1
    else:
        flag = True
        for i in range (0, len(check)):
            if check[i]<=0:
                pass
            else:
                flag = False
        if flag:
            return 2
        else:
            return 3
    
def pareto(sigma_vectors):
    vectors_matrix = np.array(sigma_vectors)
    result_matrix = [[0]*20 for i in range(20)]
    for i in range(0, 20):
        result_matrix[i][i] = 1
    for i in range(0, 20):
        for j in range(i+1, 20):
            flag = pareto_check_vector(vectors_matrix[i][j])
            if flag == 1:
                result_matrix[i][j] = 1
                result_matrix[j][i] = 0
            elif flag == 2:
                result_matrix[i][j] = 0
                result_matrix[j][i] = 1
            elif flag == 3:
                result_matrix[i][j] = 0
                result_matrix[j][i] = 0
    m, sm = dom_max_r(result_matrix)
    if len(m)>0:        
        print("Найбільші за R: {}".format(m))
        print("Строго найбільші за R: {}".format(sm))
    else:
        m, sm = block_max_r(result_matrix)
        print("Максимальні за R: {}".format(m))
        print("Строго максимальні за R: {}".format(sm))
    return result_matrix

pareto(s_v)
```

#### Мажоритарне відношення

```python
def majority(sigma_vectors):
    vectors_matrix = np.array(sigma_vectors)
    result_matrix = [[0]*20 for i in range(20)]
    for i in range(0,len(vectors_matrix)):
        for j in range (i,len(vectors_matrix)):
            if vectors_matrix[i][j].sum()>0:
                result_matrix[i][j]=1
                result_matrix[j][i]=0
            elif vectors_matrix[i][j].sum()<0:
                result_matrix[i][j]=0
                result_matrix[j][i]=1
            elif vectors_matrix[i][j].sum()==0:
                result_matrix[i][j]=0
                result_matrix[j][i]=0
    if len(dom_max_p(result_matrix))>0:
        print("Найбільший за Р: {}".format(dom_max_p(result_matrix)))
    else:
        print("Максимальні за Р: {}".format(block_max_p(result_matrix)))
    return result_matrix

majority(s_v)
```

#### Лексикографічне відношення

```python
#k5>k6>k4>k3>k9>k2>k8>k1>k12>k7>k10>k11

def sort_strong_order(a):
    ordered = np.arange(len(a))
    order = np.array([ 4, 5, 3, 2, 8, 1, 7, 0, 11, 6, 9, 10])
    for i in range (0, len(a)):
        ordered[i]=a[order[i]]
    return ordered

def sort_sigma_vectors(sigma_vectors):
    vector_matrix = np.array(sigma_vectors)
    sorted_matrix = [[0]*20 for i in range(20)]
    for i in range(0,len(vector_matrix)):
        for j in range(0,len(vector_matrix)):
            sorted_matrix[i][j] = sort_strong_order(vector_matrix[i][j])
    return sorted_matrix

def lexicographic(matrix):
    sorted_vectors = np.array(matrix)
    result_matrix = [[0]*20 for i in range(20)]
    for i in range (0, len(result_matrix)):
        result_matrix[i][i]=0
    for i in range(0, len(sorted_vectors)):
        for j in range(i+1, len(sorted_vectors)):
            a = sorted_vectors[i][j]
            for k in range(0, len(a)):
                if a[k]==1:
                    result_matrix[i][j]=1
                    result_matrix[j][i]=0
                    break
                elif a[k]==-1:
                    result_matrix[i][j]=0
                    result_matrix[j][i]=1
                    break
                else:
                    continue
    if len(dom_max_p(result_matrix))>0:
        print("Найбільший за Р: {}".format(dom_max_p(result_matrix)))
    else:
        print("Максимальні за Р: {}".format(block_max_p(result_matrix)))
    return result_matrix

lexicographic(sort_sigma_vectors(s_v))
            
```
